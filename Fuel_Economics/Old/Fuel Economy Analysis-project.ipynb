```python
for i in range(len(df_alpha18[df_alpha18['Fuel'].str.contains('/')])):
    cmlist = df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['City_MPG'].split('/')
    hmlist = df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Hwy_MPG'].split('/')
    blist = df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Cmb_MPG'].split('/')
    flist = df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Fuel'].split('/')
    df_alpha18.append({
        'Model': df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Model'], 
        'Displ': df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Displ'], 
        'Cyl': df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Cyl'], 
        'Trans': df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Trans'], 
        'Drive': df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Drive'], 
        'Fuel': flist[1],
        'Sales_Area': df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Sales_Area'],
        'Stnd': df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Stnd'],
        'Underhood_ID': df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Underhood_ID'],
        'Veh_Class': df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Veh_Class'],
        'Air_Pollution_Score': df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Air_Pollution_Score'],
        'City_MPG': cmlist[1],
        'HWY_MPG': hmlist[1],
        'CMB_MPG': blist[1],
        'Greenhouse_Gas_Score': df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Greenhouse_Gas_Score']
    }, ignore_index=True)
    df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['City_MPG'] = flist[0]
    df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Hwy_MPG'] = hmlist[0]
    df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Cmb_MPG'] = blist[0]
    df_alpha18[df_alpha18['Fuel'].str.contains('/')].iloc[i]['Fuel'] = flist[0]
    
```
